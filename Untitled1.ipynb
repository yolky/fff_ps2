```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k1 = 0.05
d1 = 0.11
k2 = 5
d2 = 1
k3 = 20
d3 = 0.1

n_timesteps = 20000
```

```python
promoter_states = np.zeros(n_timesteps + 1)
mrna_counts = np.zeros(n_timesteps + 1)
protein_counts = np.zeros(n_timesteps + 1)
time_points = np.zeros(n_timesteps + 1)


for t in range(n_timesteps):
    TF_on = k1 * (promoter_states[t] == 0)
    TF_off = d1 * (promoter_states[t] == 1)
    r_make = k2 * (promoter_states[t] == 1)
    r_deg = d2 * mrna_counts[t]
    p_make = k3 * mrna_counts[t]
    p_deg = d3 * protein_counts[t]
    
    r_tot = TF_on + TF_off + r_make + r_deg + p_make + p_deg
    
    probs = [TF_on/r_tot, TF_off/r_tot, r_make/r_tot, r_deg/r_tot, p_make/r_tot, p_deg/r_tot]
    
    
    reaction_choice = np.random.choice(list(range(6)), p = probs)
    
    promoter_states[t+1] = promoter_states[t]
    mrna_counts[t+1] = mrna_counts[t]
    protein_counts[t+1] = protein_counts[t]
    
    if reaction_choice == 0:
        promoter_states[t+1] = 1
    elif reaction_choice == 1:
        promoter_states[t+1] = 0
    elif reaction_choice == 2:
        mrna_counts[t+1] = mrna_counts[t] + 1
    elif reaction_choice == 3:
        mrna_counts[t+1] = mrna_counts[t] - 1
    elif reaction_choice == 4:
        protein_counts[t+1] = protein_counts[t] + 1
    elif reaction_choice == 5:
        protein_counts[t+1] = protein_counts[t] - 1
        
    time_diff = np.random.exponential(1/r_tot)
    
    time_points[t+1] = time_points[t] + time_diff
```

```python
plt.figure(dpi = 150)
plt.step(time_points, promoter_states)
plt.title('Promoter State')
plt.xlabel('Time (s)')
```

```python
plt.figure(dpi = 150)
plt.step(time_points, mrna_counts)

plt.title('mRNA Count')
plt.xlabel('Time (s)')
```

```python
plt.figure(dpi = 150)
plt.step(time_points, protein_counts)

plt.title('Protein Count')
plt.xlabel('Time (s)')
```

```python
k = 1.5
d = .1
K = 2
n = 6

n_timesteps = 2000
```

```python
states = np.zeros([n_timesteps + 1, 3])
time_points = np.zeros(n_timesteps + 1)

states[0, 0] = 10
```

```python
for t in range(n_timesteps):
    rates = []
    for i in range(3):
        rates.append(k/(1 + (states[t, (i - 1)%3]/K)**n))
    for i in range(3):
        rates.append(d * states[t, i])
        
    
    probs = np.array(rates)/np.sum(rates)
    
    reaction_choice = np.random.choice(list(range(6)), p = probs)
    
    states[t+1] = states[t]
    
    if reaction_choice <= 2:
        states[t+1, reaction_choice] = states[t+1, reaction_choice] + 1
    else:
        states[t+1, reaction_choice - 3] = states[t+1, reaction_choice - 3] - 1
        
    time_diff = np.random.exponential(1/r_tot)
    
    time_points[t+1] = time_points[t] + time_diff
```

```python
plt.figure(dpi = 150, figsize = (10, 4))
plt.step(time_points, states[:, 0])
plt.step(time_points, states[:, 1])
plt.step(time_points, states[:, 2])

plt.title('Repressilator Simulation, n = {}'.format(n))
plt.ylabel('mRNA Count')
plt.legend(['mRNA 1', 'mRNA 2', 'mRNA 3'])
```
