```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import sklearn.linear_model
```

# Setting Diffusion Parameters

```python
delta_t = 0.01
k_on = 0.05
k_off = 0.45

d_free = 1
d_bound = 0

p_on = 1 - np.exp(-1 * k_on * delta_t)
p_off = 1 - np.exp(-1 * k_off * delta_t)


k_on_damaged = 4.5
p_on_damaged = 1 - np.exp(-1 * k_on_damaged * delta_t)
```

# Functions for updating and initializing states

```python
def update_positions(positions, diffusion_coefficients, delta_t, min_pos = [-np.inf, -np.inf], max_pos = [np.inf, np.inf]):
    #updates positions given current positions, diffusion coefficients of each particle and time difference
    
    noise = np.random.normal(size = positions.shape)
    displacements = np.sqrt(2 * diffusion_coefficients * delta_t) * noise
    positions = positions + displacements
    
    min_pos = np.array(min_pos)
    max_pos = np.array(max_pos)
    
    #reflect off of walls if we have min/max positions
    if min_pos[0] != -np.inf:
        positions = np.abs(positions - min_pos[None, :]) + min_pos[None, :]
    if max_pos[0] != np.inf:
        positions = -np.abs(positions - max_pos[None, :]) + max_pos[None, :]
    
    return positions

def update_states_and_diffusions(states, p_on, p_off, d_on, d_off):
    #updated bound/unbound states and diffusion coefficients given transition probabilities
    
    turned_on = np.random.uniform(size = states.shape)
    turned_off = np.random.uniform(size = states.shape)
    
    states_new = (1 * (turned_on < p_on) * (states == 0)) + ((turned_off > p_off) * (states == 1))
    
    diffusions = d_off * states_new + (d_on * (1 - states_new))
    
    return states_new, diffusions
```

```python
def init_positions(n_particles, min_pos = [-np.inf, -np.inf], max_pos = [np.inf, np.inf], init_zero = True):
    #initializing positions for n particles
    #either at 0 or uniformly distributed within box
    
    min_pos = np.array(min_pos)
    max_pos = np.array(max_pos)
    
    if init_zero:
        return np.zeros([n_particles, 2])
    
    else:
        noise = np.random.uniform(size = [n_particles, 2])
        
        return (max_pos - min_pos)[None, :] * noise + min_pos[None, :]
    
```

# Question 1.2

```python
#init the states

n_particles = 1
positions_0 = init_positions(n_particles, init_zero = True)
diffusions_0 = np.array([[1]])
states_0 = np.array([[0]]) #1 = bound

timesteps = 1000
positions = np.zeros([timesteps + 1, n_particles, 2])

positions[0] = positions_0

time_range = np.linspace(0, timesteps * delta_t, timesteps + 1)

for t in range(timesteps):
    positions[t+1] = update_positions(positions[t], diffusions_0, delta_t)
    
    
plt.figure(dpi = 200)

msds = np.mean(np.sum(positions[:, :, :]**2, -1), 1)
plt.plot(time_range, msds)
plt.title('Mean square distance (n = 1)')
plt.ylabel('MSD (um^2)')
plt.xlabel('time (s)')

model = sklearn.linear_model.LinearRegression(fit_intercept = False)
model.fit(time_range.reshape(-1, 1), msds.reshape(-1, 1))

slope = model.coef_[0, 0] #this is equal to 4D in two dimensions


D_fit = slope/4

print('Estimate of D: {}'.format(D_fit))

plt.plot(time_range, slope * time_range)
```

```python
timesteps = 3000
delta_t = 0.01
positions = np.zeros([timesteps + 1, n_particles, 2])
states = np.zeros([timesteps + 1, n_particles, 1])
diffusions = np.zeros([timesteps + 1, n_particles, 1])

positions[0] = positions_0
states[0] = states_0
diffusions[0] = diffusions_0

time_range = np.linspace(0, timesteps * delta_t, timesteps + 1)
```

```python
for t in range(timesteps):
    positions[t+1] = update_positions(positions[t], diffusions[t], delta_t, min_pos = [-4, -2], max_pos = [4, 2])
    
    damaged_indices = np.where(np.logical_and(positions[t, :, 0]<=0.5, positions[t, :, 0]>=-0.5))[0]
    undamaged_indices = np.where(np.logical_not(np.logical_and(positions[t, :, 0]<=0.5, positions[t, :, 0]>=-0.5)))[0]
    
    new_states_damaged, new_diffusions_damaged = update_states_and_diffusions(states[t][damaged_indices], p_on_damaged, p_off, d_free, d_bound)
    new_states_undamaged, new_diffusions_undamaged = update_states_and_diffusions(states[t][undamaged_indices], p_on, p_off, d_free, d_bound)
    
    states[t+1][damaged_indices] = new_states_damaged
    diffusions[t+1][damaged_indices] = new_diffusions_damaged
        
    states[t+1][undamaged_indices] = new_states_undamaged
    diffusions[t+1][undamaged_indices] = new_diffusions_undamaged
```
